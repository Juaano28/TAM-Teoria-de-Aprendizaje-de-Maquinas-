```python
# Probabilidades iniciales de seleccionar cada caja
prob_roja = 0.4
prob_azul = 0.6

# Frutas en cada caja (manzanas y naranjas)
manzanas_roja = 2
naranjas_roja = 6
manzanas_azul = 3
naranjas_azul = 1

# Total de frutas por caja
total_roja = manzanas_roja + naranjas_roja  # 8
total_azul = manzanas_azul + naranjas_azul  # 4

# -----------------------------------------------------------------
# 1. Calcular P(F = A): Probabilidad de seleccionar una manzana
# -----------------------------------------------------------------
prob_A_dado_roja = manzanas_roja / total_roja  # 2/8 = 0.25
prob_A_dado_azul = manzanas_azul / total_azul  # 3/4 = 0.75

# Ley de probabilidad total
prob_A = (prob_roja * prob_A_dado_roja) + (prob_azul * prob_A_dado_azul)

# -----------------------------------------------------------------
# 2. Calcular P(F = O): Probabilidad de seleccionar una naranja
# -----------------------------------------------------------------
prob_O_dado_roja = naranjas_roja / total_roja  # 6/8 = 0.75
prob_O_dado_azul = naranjas_azul / total_azul  # 1/4 = 0.25

# Ley de probabilidad total
prob_O = (prob_roja * prob_O_dado_roja) + (prob_azul * prob_O_dado_azul)

# -----------------------------------------------------------------
# 3. Calcular P(B = r | F = O): Probabilidad de caja roja dado que se obtuvo una naranja
# -----------------------------------------------------------------
# Teorema de Bayes
prob_roja_dado_O = (prob_O_dado_roja * prob_roja) / prob_O

# -----------------------------------------------------------------
# 4. Calcular P(B = b | F = A): Probabilidad de caja azul dado que se obtuvo una manzana
# -----------------------------------------------------------------
# Teorema de Bayes
prob_azul_dado_A = (prob_A_dado_azul * prob_azul) / prob_A

# 5. Calcular P(B = b | F = O)
prob_azul_dado_O = (prob_O_dado_azul * prob_azul) / prob_O

# 6. Calcular P(B = r | F = A)
prob_roja_dado_A = (prob_A_dado_roja * prob_roja) / prob_A
# -----------------------------------------------------------------
# Resultados
# -----------------------------------------------------------------
print(f"1. P(F = A) = {prob_A:.2f} ({prob_A*100:.1f}%)")
print(f"2. P(F = O) = {prob_O:.2f} ({prob_O*100:.1f}%)")
print(f"3. P(B = r | F = O) = {prob_roja_dado_O:.4f} ({prob_roja_dado_O*100:.2f}%)")
print(f"4. P(B = b | F = A) = {prob_azul_dado_A:.4f} ({prob_azul_dado_A*100:.2f}%)")
print(f"5. P(B = b | F = O) = {prob_azul_dado_O:.4f} ({prob_azul_dado_O*100:.2f}%)")
print(f"6. P(B = r | F = A) = {prob_roja_dado_A:.4f} ({prob_roja_dado_A*100:.2f}%)")
# -----------------------------------------------------------------
```

```python
def main():
    # Pedir datos al usuario
    boxes = []
    fruits = set()

    num_boxes = int(input("¿Número de cajas? "))

    # Validar que las probabilidades sumen 1
    total_prob = 0.0

    for i in range(num_boxes):
        print(f"\nCaja {i + 1}:")
        name = input("  Nombre: ")
        prob = float(input("  Probabilidad (0-1): "))
        total_prob += prob

        # Almacenar frutas y cantidades
        box_fruits = {}
        while True:
            fruta = input("  Fruta (dejar vacío para terminar): ").strip().lower()
            if not fruta:
                break
            cantidad = int(input(f"  Cantidad de {fruta}: "))
            box_fruits[fruta] = cantidad
            fruits.add(fruta)

        # Calcular total de frutas en la caja
        total = sum(box_fruits.values())
        if total == 0:
            print("¡Error! La caja debe contener frutas.")
            return

        boxes.append({
            "nombre": name,
            "prob": prob,
            "frutas": box_fruits,
            "total": total
        })

    # Validar suma de probabilidades
    if not abs(total_prob - 1.0) < 1e-9:
        print("¡Error! Las probabilidades deben sumar 1")
        return

    # Convertir a lista ordenada
    fruits = sorted(fruits)

    # Calcular P(F) para cada fruta
    p_f = {fruta: 0.0 for fruta in fruits}
    for fruta in fruits:
        for box in boxes:
            count = box["frutas"].get(fruta, 0)
            p_f[fruta] += box["prob"] * (count / box["total"])

    # Calcular P(B|F) para cada combinación
    resultados = []
    for fruta in fruits:
        if p_f[fruta] == 0:
            continue  # Evitar división por cero

        for box in boxes:
            count = box["frutas"].get(fruta, 0)
            numerador = (count / box["total"]) * box["prob"]
            p_b_dado_f = numerador / p_f[fruta]
            resultados.append((box["nombre"], fruta, p_b_dado_f))

    # Mostrar resultados
    print("\nRESULTADOS:")
    print("\nProbabilidades marginales de las frutas:")
    for fruta in fruits:
        print(f"P({fruta.upper()}) = {p_f[fruta]:.4f} ({p_f[fruta]*100:.2f}%)")

    print("\nProbabilidades condicionales de las cajas:")
    for box, fruta, prob in resultados:
        print(f"P({box.upper()} | {fruta.upper()}) = {prob:.4f} ({prob*100:.2f}%)")

if __name__ == "__main__":
    main()
```
